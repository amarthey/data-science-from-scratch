# 03_visualizing_data

```python
import matplotlib.pyplot as plt
from collections import Counter

# %matplotlib inline
```

```python
%matplotlib notebook

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('Nominal GDP')
plt.ylabel('Billions of $')
# plt.show()
```

```python
%matplotlib notebook

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

x = [i for i, _ in enumerate(num_oscars)]

plt.bar(x, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.xticks([i + 0.5 for i, _ in enumerate(num_oscars)], movies)
plt.show()
```

```python
%matplotlib notebook

scores = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda x: x // 10 * 10

grade_counts = Counter(decile(score) for score in scores)
grades = [x -4 for x in grade_counts.keys()]
counts = [x for x in grade_counts.values()]

plt.bar(left=grades, height=counts, width=8)
plt.axis([-5, 105, 0, 5])
plt.xticks([i for i in range(0, 110, 10)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
%matplotlib notebook

mentions = [500, 505]
years = [2013, 2014]

plt.bar(left=[2012.6, 2013.6], height=mentions, width=0.8)
plt.xticks(years)

plt.ylabel("# of times I heard someone say 'data science'")
plt.ticklabel_format(useOffset=False)

plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
%matplotlib notebook

plt.bar(left=[2012.6, 2013.6], height=mentions, width=0.8)
plt.xticks(years)

plt.ylabel("# of times I heard someone say 'data science'")
plt.ticklabel_format(useOffset=False)

plt.axis([2012.5, 2014.5, 0, 550])
plt.title("Not So Huge Anymore.")       
plt.show()
```

```python
%matplotlib notebook

variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]

xs = range(len(variance))

plt.plot(xs, variance,     'g-',  label='variance')    # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')      # red dot-dashed line
plt.plot(xs, total_error,  'b:',  label='total error') # blue dotted line

plt.legend(loc=9) # loc=9 means "top center"
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
%matplotlib notebook

friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend, minute in zip(labels, friends, minutes):
    plt.annotate(label, 
                 xy=(friend, minute), 
                 xytext=(5, -5), # slightly offset
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

```python
%matplotlib notebook

test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

axes[0].scatter(test_1_grades, test_2_grades)
axes[0].set_xlabel('test 1 grade')
axes[0].set_ylabel('test 2 grade')
axes[0].set_title("Axes Are Comparable")
axes[0].axis("equal")

axes[1].scatter(test_1_grades, test_2_grades)
axes[1].set_xlabel('test 1 grade')
axes[1].set_ylabel('test 2 grade')
axes[1].set_title("Axes Are Comparable")
```
